```python
# N % 2 == 0
s = [x for x in range(1, 11)]
median = (s[int(len(s)/2 - 1)]+s[int(len(s)/2) + 1 - 1])/2 if len(s) % 2 == 0 else s[int((len(s)+1)/2 - 1)]
print(f"Median: {median}")
```

```python
# N % 2 != 0
s = [x for x in range(1, 12)]
median = (s[int(len(s)/2 - 1)]+s[int(len(s)/2) + 1 - 1])/2 if len(s) % 2 == 0 else s[int((len(s)+1)/2 - 1)]
print(f"Median: {median}")
```

```python
list_odd_filter = [x for x in s if x % 2 != 0]
list_odd_filter.sort(reverse=True)
print(list_odd_filter)
```
