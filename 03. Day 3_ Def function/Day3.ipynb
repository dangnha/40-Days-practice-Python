**Objective**: If/else, def function

```python
def quadratic_equation(a, b, c):
  x1, x2 = 0, 0
  denta = b**2 - 4*a*c
  if denta > 0:
      x1 = (-b + denta**0.5)/(2*a)
      x2 = (-b - denta**0.5)/(2*a)
      print(f"x1 = {x1}, x2 = {x2}")
  if denta == 0:
      x1 = -b/2*a
      x2 = -b/2*a
      print(f"x1 = x2 = {x2}")
  if denta < 0:
      print("The equation don't have valid x1, x2")
```

```python
quadratic_equation(2,6,4)
```

```python
quadratic_equation(1,2,1)
```

```python
quadratic_equation(4,6,3)
```
