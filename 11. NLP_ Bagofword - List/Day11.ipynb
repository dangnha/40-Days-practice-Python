# Bag of Words  
Bag of words is a algorithms preprocessing raw data in Natural Language Processing. The idea of BOW is alalysis and group based on "Bag of Words" (corpus).
* Step 1: Split document into single word
* Step 2: Create a set of single word in the document ad they don't duplicate
* Step 3: Present 1 input sentence by vector, each element represents the number of time this word appers in the input sentence.  
image.png

```python
corpus = ["Tôi thích môn toán", "Tôi thích AI", "Tôi thích âm nhạc"]

def tokenize(corpus):
    rawList = []
    # Split text in corpus
    for text in corpus:
        splitList = list(map(str, text.split()))
        rawList += splitList
    # Remove duplicate token by using set data structure
    print(f"Raw List: {rawList}")
    uniqueList = set()
    for x in rawList:
        uniqueList.add(x)
    print(f"Bag of Words: {uniqueList}")
    return list(uniqueList)
BOW = tokenize(corpus)

# Vectorize data
def vectorize(data):
    dataToken = []
    for text in data:
          splitList = list(map(str, text.split()))
          dataToken += splitList
    print(f"Token in data: {dataToken}")
    vector = [dataToken.count(x) for x in BOW]
    return vector
data = ["Tôi thích AI thích toán"]
print(vectorize(data))
```
