# String trong python  

## Lý thuyết

```python
s = "hello world"
s
```

```python
'%s %s %d' % ("hello", "world", 2024)
```

```python
# Build-in function
print(s.capitalize())
print(s.upper())
print(s.lower())
print(s.rjust(7))
print(s.center(7))
print(s.title())
print(s.replace("llo", "(hehe)"))
print("  -nha-  ".strip())
```

```python
# String format
name = "Dang Nha"
age = 21
infor = f"Có một người tên {name} tuổi {age}"
infor
```

```python
# String format handle floats
val = 12.3
print(f'{val:.2f}') # Returns 12.30
print(f'{val:.5f}') # Returns 12.30000
```

```python
# Use special character in string
name = 'his name is \'dangnha\''
name
```

## Bài tập:  


```python
"""
Bài 1: Tạo một biến name gán giá trị là tên một người, sau đó hiển thị ra
màn hình một thông báo chưa tên đó sử dụng f-string.
"""
name = "dangnha"
print(f"His name is {name}")
```

```python
"""
Bài 2: Tạo một biến và giá trị "ms Taylor" cho nó, sau đó thực hiện in ra
màn hình giá trị biến đó được viết hoa kí tự đầu tiên mỗi từ hoặc toàn bộ
đều là chữ in hoa, hoặc chữ thường
"""
name = "ms Taylor"
print(name.title())
print(name.upper())
print(name.lower())
```
