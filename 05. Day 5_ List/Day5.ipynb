```python
"""
Câu 1: Tạo mới 1 List gồm các số từ 1 tới 10
"""
ds = [x for x in range(1, 11)]
ds
```

```python
"""
Câu 2: In ra kết quả 5 phần tử đầu tiên có trong List trên
"""
ds[:5]
```

```python
"""
Câu 3: In ra kết quả phần tử có giá trị không chia hết cho 2
"""
new_ds = [x for x in ds if x % 2 == 1]
new_ds
```

```python
"""
Câu 4: In ra kết quả tổng các giá trị trong list
"""
print(sum(ds))
```
