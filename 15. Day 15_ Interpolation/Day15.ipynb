# Interpolation
 **Interpolation** is a mathematical technique that estimates the values of unknown data points between known data points.  
 We have some methodology such as Nearest Neighbor and Linear Interpolation.

# Exercise:
Use the method Nearest Neighbor:
* Step 1: Find the point needt to apply interpolation for predicting value
* Step 2: Fill the value based on the neaghbor data
image.png

```python
def nearestNeighbor(data):
    for i, x in enumerate(data):
        l, r = -1, -1
        if x == None:
            for m in range(i, -1, -1): # Find the left position
                if data[m] != None:
                    l = m
                    break
            for n in range(i, len(data)-1): # Find the right position
                if data[n] != None:
                    r = n
                    break
            if l == -1: # It mean in the left position we don't have the valid value
                data[i] = data[r]
            elif r == -1: # It mean in the right position we don't have the valid value
                data[i] = data[l]
            elif l != -1 and r != -1: # We've found valid value in left and right
                data[i] = round((data[l] + data[r]) / 2, 2)
            else: # All data cell are None
                print("The dataset is empty")
                return -1
    return data
```

## Normal testcase

```python
lst_data1 = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
nearestNeighbor(lst_data1)
```
