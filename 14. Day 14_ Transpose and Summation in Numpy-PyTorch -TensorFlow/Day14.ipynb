# Transpose theory
**Transpose** is a mathemathic methodology in linear algebra. It turn the position of each element in rows into the position of each element in columns and vice versa.
$$(A^T)_{mn} \; = \; A_{nm}$$
image.png

## Properties

````python
import numpy as np
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
````

### $A = (A^T)^T$

````python
print(f"A = {A}")
print(f"(A.T).T = {(A.T).T}")
````

### $(A+B)^T = A^T + B^T$

````python
print(f"A.T + B.T = {A.T + B.T}")
print(f"(A+B).T = {(A+B).T}")
````

### $(A*B)^T = B^T * A^T$

````python
print(f"(A*B).T = {(A*B).T}")
print(f"B.T * A.T = {B.T * A.T}")
````

## Transpose in Pytorch and Tensorflow

````python
import torch
A = torch.tensor([[1,2], [3,4]])
tensor_transpose1 = torch.t(A)
tensor_transpose2 = torch.transpose(A, 1, 0)
print(f"Torch.t transpose: {tensor_transpose1}")
print(f"Torch.transpose transpose: {tensor_transpose2}")
````

````python
import tensorflow as tf
A = torch.tensor([[1,2], [3,4]])
tensor_transpose = tf.transpose(A)
print(f"Tf transpose: {tensor_transpose}")
````

# Summation theory
In numpy, np.sum() is a function that's used for calculating the sum of all elements in an array.  
```python3
np.sum(a, axis=None, dtype=None, keepdims=False, initial=0, where=True)
```
Where:  
* a: Input array
* axis: (optional) The dimension in the array will be choosed for take the summation
* dtype: (optional) The datatype of the result
* keepdims: (optional) If it's True, keep the dimension of result array equal dimension of input array
* initial: (optional) The initial value for summation
* where: (optional) A Boolean array specifies the positions of the elements used in the operation.

## Sum in numpy

````python
# Define 2D array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Sum of all elements in array
total_sum = np.sum(arr)

# Calculate totals by column
column_sum = np.sum(arr, axis=0)

# Calculate totals by row
row_sum = np.sum(arr, axis=1)

print("Array :\n", arr)
print("Sum of all elements in array:", total_sum)
print("Totals by column:", column_sum)
print("Totals by row:", row_sum)
````

## Sum in Pytorch

````python
# Create 2d tensor
tensor_1 = torch.tensor([[1, 2], [3, 4]])

# Calculate the tensor sum
tensor_sum = torch.sum(tensor_1)

# Calculate the sum of tensors according to column axis = 0
tensor_sum_axis_0 = torch.sum(tensor_1, axis=0)

# Calculate the sum of tensors along axis = 1
tensor_sum_axis_1 = torch.sum(tensor_1, axis=1)

print(" tensor :\n", tensor_1)
print("Sum of elements in tensor: \n", tensor_sum)
print("Sum of elements by column: \n", tensor_sum_axis_0)
print("Sum of elements in rows: \n", tensor_sum_axis_1)
````

## Sum in Tensorflow

````python
# Create 2d tensor
tensor_1 = tf.constant([[1, 2], [3, 4]])

# Calculate the tensor sum
tensor_sum = tf.reduce_sum(tensor_1)

# Calculate the sum of tensors according to column axis = 0
tensor_sum_axis_0 = tf.reduce_sum(tensor_1, axis=0)

# Calculate the sum of tensors along axis = 1
tensor_sum_axis_1 = tf.reduce_sum(tensor_1, axis=1)

print(" tensor :\n", tensor_1)
print("Sum of elements in tensor: \n", tensor_sum)
print("Sum of elements by column: \n", tensor_sum_axis_0)
print("Sum of elements in rows: \n", tensor_sum_axis_1)
````
