# Define array function in Numpy, Tensorflow, Pytorch

```python
import numpy as np
import tensorflow as tf
import torch
```

## Zeros

```python
zeros_np = np.zeros((2,3))
zeros_np
```

```python
zeros_pt = torch.zeros((3,3))
zeros_pt
```

```python
zeros_tf = tf.zeros((3,4))
zeros_tf
```

## Ones

```python
ones_np = np.ones((2,3))
ones_np
```

```python
ones_pt = torch.ones((3,3))
ones_pt
```

```python
ones_tf = tf.ones((3,4))
ones_tf
```

## Full, fill

```python
array_np = np.full((2,3), 5)
array_np
```

```python
array_pt = torch.full((3,2), 2)
array_pt
```

```python
array_tf = tf.fill((7,5), 8)
array_tf
```
