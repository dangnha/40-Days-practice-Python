```python
"""
Câu 1: Tạo mới một List có tên là lst_data, gồm ccs số chẵn từ 1 đến 12
"""
lst_data = [x for x in range(1,13) if x % 2 == 0]
lst_data
```

```python
"""
Câu 2: Xóa tất cả các số chia hết cho 3 trong lst_data
"""
for x in lst_data:
    if x % 3 == 0:
        lst_data.remove(x)
lst_data
```

```python
"""
Câu 3: Thêm vào chuỗi lst_data các số từ 1 đến 3, và thêm vào
vị trí index = 3 một chuỗi các số từ 6 đến 8
"""
for i in range(1, 4):
    lst_data.append(i)
print(lst_data)
index = 3
for i in range(6, 9):
    lst_data.insert(index, i)
    index += 1
print(lst_data)
```

```python
"""
Câu 4: Nếu các số trong lst_data chia hết cho 2 hoặc
chia hết cho 5 thì cập nhật thành số 0
"""
for i in range(len(lst_data)):
    if lst_data[i] % 2 == 0 or lst_data[i] % 5 == 0:
        lst_data[i] = 0
print(lst_data)
```
