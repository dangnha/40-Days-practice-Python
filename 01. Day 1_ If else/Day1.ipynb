**Objective: If-else statement**
##**Can**: Lấy năm sinh chia cho 10 và lấy dư. Nếu phần dư bằng 0-9 tương ứng với các can sau:

## **Chi**: Lấy năm sinh chia cho 12 và lấy phần dư. Nếu phần dư từ 0-11 tương ứng với Chi:  
image.png


```python
def calculateCanChiCalendar(year):
    result = ''
    canDu = year % 10
    if canDu == 0:
        result += "Canh"
    elif canDu == 1:
        result += "Tân"
    elif canDu == 2:
        result += "Nhâm"
    elif canDu == 3:
        result += "Quý"
    elif canDu == 4:
        result += "Giáp"
    elif canDu == 5:
        result += "Ất"
    elif canDu == 6:
        result += "Bính"
    elif canDu == 7:
        result += "Đinh"
    elif canDu == 8:
        result += "Mậu"
    elif canDu == 9:
        result += "Kỷ"

    result += " "

    chiDu = year % 12
    if chiDu == 0:
        result += "Thân"
    elif chiDu == 1:
        result += "Dậu"
    elif chiDu == 2:
        result += "Tuất"
    elif chiDu == 3:
        result += "Hợi"
    elif chiDu == 4:
        result += "Tí"
    elif chiDu == 5:
        result += "Sửu"
    elif chiDu == 6:
        result += "Dần"
    elif chiDu == 7:
        result += "Mẹo"
    elif chiDu == 8:
        result += "Thìn"
    elif chiDu == 9:
        result += "Tỵ"
    elif chiDu == 10:
        result += "Ngọ"
    elif chiDu == 11:
        result += "Mùi"

    return result
```

```python
print(calculateCanChiCalendar(2013))
print(calculateCanChiCalendar(2003))
print(calculateCanChiCalendar(2000))
print(calculateCanChiCalendar(1971))
print(calculateCanChiCalendar(1969))
```
