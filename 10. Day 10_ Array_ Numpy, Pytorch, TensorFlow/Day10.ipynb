# Array and Tensor in Pytorch/ Tensorflow

Array in Numpy and Tensor in Pytorch, Tensorflow are fundamental units for taking the computation in those library.  
* **Array** is a data structure that help stores and performs calculation on numaric data. Array can be a scalar, vector, matrix or tensor depend on the size of data.
* **Tensor in Pytorch and tensorflow** is a data structure like array in Numpy but it's designed for dealing with deep learning and parallel calculation, expectially calculate in CPU and TPU. It is a fundamental unit for storing and handling data

```python
import numpy as np
import tensorflow as tf
import torch

listt = [4, 2, 5, 2, 4]

# Numpy
n = np.array(listt)
# Pytoch
m = torch.tensor(listt)
# Tensorflow
i = tf.constant(listt)
n, m, i
```

When we work with tensor, we can check their properties through some method such as:  
* **shape** size of tensor
* **dtype** the datatype of elemens in tensor
* **type** the type of object or tensor
* **device** (available in Pytorch and Tensorflow) help we know where the tensor is stored, CPU or GPU

## Ex 2: Write a program to create Numpy array, Tensorflow tensor, Pytorch tensor from a 2-dimensional list. Then check the shape, dtype, type, device attributes from the newly created arrays and tensors?

```python
list_2D = [[1,2,3], [4,5,6], [7,8,9]]
```

```python
array = np.array(list_2D)
print(array)
print(array.shape)
print(array.dtype)
print(type(array))
```

```python
tensor_2D_pt = torch.tensor(list_2D)
print(tensor_2D_pt)
print(tensor_2D_pt.shape)
print(tensor_2D_pt.dtype)
print(type(tensor_2D_pt))
print(tensor_2D_pt.device)
```

```python
tensor_2D_tf = tf.convert_to_tensor(list_2D)
print(tensor_2D_tf)
print(tensor_2D_tf.shape)
print(tensor_2D_tf.dtype)
print(type(tensor_2D_tf))
print(tensor_2D_tf.device)
```
